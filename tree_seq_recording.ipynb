```python
import msprime, tskit, pyslim
ts = tskit.load("/Users/ayeraselvan/Desktop/simulations/tree seq recording/overlay.trees")
ts = ts.simplify()
mutated = msprime.sim_mutations(ts, rate=1e-7, random_seed=1, keep=True)
mutated.dump("/Users/ayeraselvan/Desktop/simulations/tree seq recording/overlay_II.trees")
```
